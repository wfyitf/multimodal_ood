## Example Showing for MELD data

```python
import torch
import clip
from PIL import Image
# Load CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
clip_model, preprocess = clip.load("ViT-B/32", device=device)
df_table = pd.read_csv(f'{dataset_dir}/meldood/sample/train_annotation.csv')
```

```python

k = 1
image_id = f'dia{df_table.iloc[k]['Dialogue_ID']}_utt{df_table.iloc[k]["Utterance_ID"]}.jpg'
print('Image ID:', image_id)
print('*'*50)
print('Emotions:', df_table.iloc[k]['Emotion'])
print('Utterance:', df_table.iloc[k]['Utterance'])

print('*'*50)
plot_image(f"Image: {image_id}",
              f"{dataset_dir}/meldood/sample/train_image/{image_id}")



# Define the image path and the text
image_path = f"{dataset_dir}/meldood/sample/train_image/{image_id}"
label = ['joy', 'neutral', 'sadness', 'anger', 'fear', 'disgust', 'surprise']

# Preprocess the image
image = preprocess(Image.open(image_path)).unsqueeze(0).to(device)

print('*'*50)
for emo in label:
    text = f"The speaker is expressing {emo}."
    # Tokenize the text
    text_tokens = clip.tokenize([text]).to(device)

    # Encode the image and the text
    with torch.no_grad():
        image_features = clip_model.encode_image(image)
        text_features = clip_model.encode_text(text_tokens)

    # Normalize the features
    image_features = image_features / image_features.norm(dim=-1, keepdim=True)
    text_features = text_features / text_features.norm(dim=-1, keepdim=True)

    # Calculate the cosine similarity
    cosine_similarity = torch.matmul(image_features, text_features.T)

    # Print the cosine similarity
    print(f"Cosine Similarity for {emo}:", cosine_similarity.item())
```
