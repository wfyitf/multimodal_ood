```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json 
import sys 
```

```python
current_dir = os.getcwd()
dataset_dir = os.path.join(os.path.dirname(current_dir), 'dataset')
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def plot_image(caption, image_path):
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis('off') 
    plt.title(caption)
    plt.show()
```

## Example Showing for QA OOD detection data

```python
# Real annotation table for coco
df_table = pd.read_json(f'{dataset_dir}/qaood/sample.json')
```

```python
k = 2
print('Image ID:', df_table.iloc[k]['image_id'])
print('*'*50)
print(df_table.iloc[k]['dialog_full'])
print('*'*50)
print(df_table.iloc[k]['supercategories'])
print(df_table.iloc[k]['categories'])
print('*'*50)
plot_image(df_table.iloc[k]['caption'], 
           f"{dataset_dir}/qaood/sample/COCO_train2014_{df_table.iloc[k]['image_id']:0>12}.jpg")

```

## Example Showing for Real OOD detection data

```python
import PIL
def show_image(image, title):
    plt.imshow(PIL.Image.fromarray(image))
    plt.title(title)
    plt.show()
```

```python
df_table = pd.read_json(f'{dataset_dir}/realood/sample.json')
```

```python
k = 15
print('Image ID:', df_table.iloc[k]['image_id'])
print('*'*50)
count = 0 
for i in df_table.iloc[k]['dialog']:
    if count % 2 == 0:
        print('S1:', i)
    else:
        print('S2:', i)
    count += 1
print('*'*50)
print('Categories:', df_table.iloc[k]['categories'])
print('Super Categories:', df_table.iloc[k]['supercategories'])
print('*'*50)
plot_image(f"Image: {df_table.iloc[k]['image_id']}",
              f"{dataset_dir}/realood/sample/{df_table.iloc[k]['image_id']}.jpg")
```
